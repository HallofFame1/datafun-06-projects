## WordCloud

Import pandas, wordcloud, PILÂ  amd nunpy and matplotlib libraries

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

Import Allegheny county employees CSV dataset

```python
salaries_df = pd.DataFrame(pd.read_csv('allegheny_county_employees_salaries.csv', skipinitialspace = True))
```

Extract the first names and titles from the dataframe

```python
names = ' '.join(salaries_df['FIRST_NAME'])
titles = ' '.join(salaries_df['JOB_TITLE'])
```

Create the wordart using WordCloud

```python
wordcloud = WordCloud(width=1200, height=800, background_color='white').generate(names)
```

Display the result word art

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Create another word cloud from the job titles

```python
mask = np.array(Image.open('flag_allegheny_county.jpg'))
new_wc = WordCloud(width=1200, height=800, background_color='Chartreuse', mask=mask).generate(titles)

plt.imshow(new_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
